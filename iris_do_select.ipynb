```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("D:/dataset/iris_train.csv")
test = pd.read_csv("D:/dataset/iris_test.csv")
actual=pd.read_csv("D:/dataset/iris_actual.csv")
```

```python
train_x=train.drop('Species',axis=1)
train_y=train['Species']
test_x=test.drop('id',axis=1)
test_y=actual['Species']

model=KNeighborsClassifier(n_neighbors=5)
model.fit(train_x,train_y)
pred=model.predict(test_x)

print(accuracy_score(test_y,pred))
```

```python
iris_pred=pd.DataFrame()
iris_pred['id']=test['id']
iris_pred['Species']=pred
iris_pred
#iris_pred.to_csv("/code/iris_prediction.csv")
```
